Q4. Indexing using .loc
Using .loc, perform the following:

Display the row with index 10.

Display the Sales value of the row with index 20

```
# Display row with index 10
print("Row with index 10:")
print(df.loc[10])

# Display Sales value from row with index 20
print("\nSales value at index 20:")
print(df.loc[20, 'Sales'])

```

Q5. Indexing using .iloc

Using .iloc, display:

The first 6 rows and first 4 columns.

The Sales value from the 3rd row.

```
# First 6 rows and first 4 columns
print(df.iloc[:6, :4])

# Sales value from 3rd row
print('\nSales value from 3rd row:')
print(df.iloc[2]['Sales'])
```

Q7. Filtering Data

Filter the dataset to display:

Rows where Region = "East".

Rows where Sales > 1500.

```
# Rows where Region = East
print(df[df['Region'] == 'East'])

# Rows where Sales > 1500
print(df[df['Sales'] > 1500])
```

Q8. Multiple Condition Filtering

Display rows where:

Region is West

Sales is greater than 1200

```
# Region = West and Sales > 1200
filtered_df = df[(df['Region'] == 'West') & (df['Sales'] > 1200)]

print(filtered_df)
```
